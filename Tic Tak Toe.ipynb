```python
from IPython.display import clear_output
import time
import numpy as np

print("Welcome to Tic Tack Toe!")

user1 = input("Player 1: Do you want to be X or O? ")

def user_choice():
    
    choice = 'Wrong'

    acceptable_values = [1,2,3,4,5,6,7,8,9]

    while choice.isdigit() == False or int(choice) not in acceptable_values:
        choice = input("Please enter a number (1-9): ")
        print(choice)
        if choice.isdigit() == False:
            print("Sorry that is not a digit.")

        if int(choice) not in acceptable_values:
            print("Sorry, please enter a number in range 1-9 which has not been filled")
    
    
    return int(choice)


def print_matrix(matrix):

    print(" "+" "+" "+"|"+" "+" "+" "+"|"+" "+" "+" ")
    print(" "+matrix[0]+" "+"|"+" "+matrix[1]+" "+"|"+" "+matrix[2]+" ")
    print(" "+" "+" "+"|"+" "+" "+" "+"|"+" "+" "+" ")
    print("-----------")
    print(" "+" "+" "+"|"+" "+" "+" "+"|"+" "+" "+" ")
    print(" "+matrix[3]+" "+"|"+" "+matrix[4]+" "+"|"+" "+matrix[5]+" ")
    print(" "+" "+" "+"|"+" "+" "+" "+"|"+" "+" "+" ")
    print("-----------")
    print(" "+" "+" "+"|"+" "+" "+" "+"|"+" "+" "+" ")
    print(" "+matrix[6]+" "+"|"+" "+matrix[7]+" "+"|"+" "+matrix[8]+" ")
    print(" "+" "+" "+"|"+" "+" "+" "+"|"+" "+" "+" ")

def check_combinations(matrix,user1,user2):
    turn = 1
    game_end = False

    for i in range(0,9):
        clear_output()
        pos = user_choice()        
        if turn%2 == 1:
            matrix[pos-1] = user1
        else:
            matrix[pos-1] = user2

        new_arr = np.reshape(matrix,(3,3))
        new_arr1 = np.transpose(new_arr)
        new_arr2 = [new_arr[i][i] for i in range(0,3)]
        new_arr3 = [new_arr[i][len(new_arr)-i-1] for i in range(0,3)]#2,1,0
        # print(game_end)
        # print(new_arr)
        # print(new_arr1)
        # print(new_arr2)
        # print(new_arr3)
        for i in range(0,2):
            if " " not in new_arr[i] or " " not in new_arr1[i]:
                # print(game_end)
                if len(set(new_arr[i])) == 1 or len(set(new_arr1[i])) == 1 or len(set(new_arr2)) ==1 or len(set(new_arr3)) == 1:
                    game_end = True
                    print(f"Turn = {turn}")
                    # print(game_end)
                    break

        if game_end == True:
            break    



        if i < 8:
            print_matrix(matrix)
            time.sleep(5)

        turn+=1
        print(f"Turn = {turn}")

    if turn%2 == 0:
        print(f"Player 1 chose: {user1}")
        print(f"Player 2 chose: {user2}")
        print("Player 2 has won the game! Congratulations!")
            
    else:
        print(f"Player 1 chose: {user1}")
        print(f"Player 2 chose: {user2}")
        print("Player 1 has won the game! Congratulations!")
            


        
    print_matrix(matrix)



if user1.lower() == 'x':
    user2 = 'O'
else:
    user2 = 'X'


if user1.lower() == 'x':
    print("Player 1 will go first.")
else:
    print("Player 2 will go first.")

ready = input("Are you ready to play? Enter Yes or No.")

if ready.lower() == "yes":

    matrix = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    check_combinations(matrix,user1,user2)

else:
    print("Thank you, I hope you enjoyed!")
```
